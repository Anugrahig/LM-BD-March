```python
img=cv2.resize(img,(756,900))
```

```python
img.shape
```

```python
# starting poin (x,y)
# ending point (x,y)
# diagoinal point
# color in bgr


rect_img=cv2.rectangle(img,pt1=(70,10),pt2=(650,670),color=(0,0,255),thickness=6)
```

```python
plt.imshow(cv2.cvtColor(rect_img,cv2.COLOR_BGR2RGB))
```

```python
img=cv2.imread("./data/tristan_group.jpg")

# to fill image thickness -1 or cv2.Filled
rect_img=cv2.rectangle(img,pt1=(1110,500),pt2=(1700,1300),color=(255,0,0),thickness=cv2.FILLED)
rect_img=cv2.rectangle(img,pt1=(70,10),pt2=(650,670),color=(0,0,255),thickness=6)

plt.imshow(cv2.cvtColor(rect_img,cv2.COLOR_BGR2RGB))
```

```python
cv2.imshow("Image",rect_img)
cv2.waitKey()
cv2.destroyAllWindows()
```
