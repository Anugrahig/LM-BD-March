# Text on a Image

```python
img=cv2.imread("./data/group.jpg")
con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
txt_img=cv2.putText(con_img,text="My family",org=(50,500),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=4,color=(0,128,128),thickness=7)
```

```python
plt.imshow(txt_img)
```
