```python
img=cv2.imread("./data/group.jpg")
con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
rect_img=cv2.rectangle(con_img,pt1=(100,200),pt2=(600,500),color=(255,0,0),thickness=cv2.FILLED)
cir_img=cv2.circle(rect_img,center=(450,350),radius=150,color=(255,0,255),thickness=-1)
txt_img=cv2.putText(cir_img,"luminar",org=(110,350),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=1.5,color=(0,0,255),thickness=2)
```

```python
plt.imshow(txt_img)
```
