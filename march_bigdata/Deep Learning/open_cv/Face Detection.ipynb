```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("./data/group.jpg")
org_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
cv2.imshow("Image",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Face Detection using Harrcascade

```python
face_cascde=cv2.CascadeClassifier("./data/haarcascade_frontalface_default.xml")
```

```python
face_cascde
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("Gray Image",gray_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
faces=face_cascde.detectMultiScale(gray_img)
```

```python
faces
```

```python
plt.imshow(org_img)
```

```python
len(faces)
```

```python
for (x,y,w,h) in faces:
  cv2.rectangle(img,pt1=(x,y),pt2=(x+w,y+h),color=(0,255,0),thickness=5)
```

```python
plt.imshow(img)
```
