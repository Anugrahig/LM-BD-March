```python
import cv2
import matplotlib.pyplot as plt
import pytesseract
from pytesseract import Output
import re
```

```python
img=cv2.imread("./data/invoice-sample.jpg")
```

```python
cv2.imshow("Invoice",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img.shape
```

```python
# 70 % box
# date only
```

```python
img_text=pytesseract.image_to_data(img,output_type=Output.DICT)
```

```python
img_text.keys()
```

```python
date_pattern = '^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[012])/(19|20)\d\d$'
mail_pattern =	'^[a-z0-9](\.?[a-z0-9]){5,}@example\.net$'
```

```python
img=cv2.imread("./data/invoice-sample.jpg")
for i in range(len(img_text["text"])):
  if(img_text["conf"][i])>=50:
    if re.match(mail_pattern,img_text["text"][i]) or re.match(date_pattern,img_text["text"][i])  :
      x,y,w,h=img_text["left"][i],img_text["top"][i],img_text["width"][i],img_text["height"][i]
      cv2.rectangle(img,(x,y),(x+w,h+y),(0,255,0),4)
    
cv2.imshow("Invoice",img)
cv2.waitKey()
cv2.destroyAllWindows()
```
