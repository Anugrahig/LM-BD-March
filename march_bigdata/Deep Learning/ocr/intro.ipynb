```python
dark_img=cv2.imread("./data/dark_image.jpg")
```

```python
text1=pytesseract.image_to_data(dark_img,output_type=Output.DICT)
```

```python
text1.keys()
```

```python
for i in range(len(text1["text"])):
  if text1["conf"][i]>90:
    x,y,w,h=text1["left"][i],text1["top"][i],text1["width"][i],text1["height"][i]
    print(text1["left"][i])
  cv2.rectangle(dark_img,(x,y),(x+w,y+h),(0,0,255),4)
cv2.imshow("Text Image",text_img)
cv2.waitKey()
cv2.destroyAllWindows()
  
```
