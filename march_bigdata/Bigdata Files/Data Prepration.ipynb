### Handling missing values

**Dropna : dropna will delete all NaN rows**

```python
df.dropna(inplace=True)
df.fillna(300,inplace=True)
```

**Fill missing value into specific column**

```python
df["Calories"].fillna(300,inplace=True)
df.head(0)
df["Date"].fillna("2020/12/02",inplace=True)
```

**Fill missing value into specific column basis (mean,median,mode:most repeated value)**

**Object data filled with mode value**

```python
# fill NaN value with mean value
df["Calories"].fillna(df["Calories"].mean(),inplace=True)

# fill NaN value with median value
df["Calories"].fillna(df["Calories"].median(),inplace=True)
# fill NaN value with mode value
df["Calories"].fillna(df["Calories"].mode()[0],inplace=True)
```

**Fill missing value into specific column**

```python
df.dropna(subset=["Date"],inplace=True)
```

**Handle wrong formatted data**

```python
df["Calories"]=df["Calories"].apply(lambda x: round(df["Calories"].mean(),2) if x>400 else x)
```

**Delete Dulpicate values**

```python
df.drop_duplicates(inplace=True)
```
