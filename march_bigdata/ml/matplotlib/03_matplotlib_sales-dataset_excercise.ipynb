```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../../Bigdata Files/company_sales_data.csv")
df
```

**1. Read Total profit of all months and show it using a line plot**

```python
x=df.month_number
y=df.total_profit
plt.plot(x,y)
```

**2. Get total profit of all months and show line plot with the following Style properties**

 **Line Style dotted and Line-color should be red <br>
 Show legend at the lower right location.<br>
 X label name = Month Number<br>
 Y label name = Sold units number<br>
 Add a circle marker.<br>
 Line marker color as read<br>
 Line width should be 3**

```python
plt.plot(x,y,ls="--",c="red",marker="o",lw=3,label="Company Profit")
labels=df.month_number
plt.xlabel("Month Number")
plt.ylabel("Solid units number")
plt.legend(loc="lower right")
```

**3.Read all product sales data and show it  using a multiline plot
    square marker with color=green**

```python
y=['facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo', 'moisturizer']
plt.plot(df.month_number,df.facecream,df.month_number,df.facewash,\
         df.month_number,df.toothpaste,df.month_number,df.bathingsoap, \
         df.month_number,df.shampoo,df.month_number,df.moisturizer,c="g",marker="s")
plt.xlabel("Month")
plt.ylabel("Sales data")
plt.title("Month Vs. Salesdata")
plt.show()
```

**4.Read toothpaste sales data of each month and show it using a scatterplot**

```python
plt.scatter(df.month_number,df.toothpaste)
plt.xlabel("Month Number")
plt.ylabel("Toothpaste")
plt.title("Toothpaste Salesdata")
plt.grid(linestyle='--', linewidth=1)
```

**5.Read face cream and facewash product sales data and show it using the bar chart with grid line**

```python
plt.bar(df.month_number,df.facecream,label="facecream")
plt.bar(df.month_number,df.facewash,label="facewash")
plt.grid(color="c")
plt.xlabel("Month Number")
plt.ylabel("Facecream , Facewash")
plt.title("Month vs Facecream,Facewash")
plt.legend()
```

**6.Read the total profit of each month and show it using the histogram to see the most common profit ranges**

```python
plt.hist(df.total_profit,label="profit data")
plt.xticks([240000,280000,320000,360000,400000])
plt.xlabel("total Profit")
plt.legend()
plt.show()
```

**7.Calculate total sale data for last year for each product and show it using a Pie chart**

```python
x=[sum(df.facecream),sum(df.facewash),sum(df.toothpaste),sum(df.bathingsoap),sum(df.shampoo),sum(df.moisturizer)]
labels=["Facecream","Facwash","Toothpaste","Bathingsoap","Shampoo","Moisturizer"]
plt.pie(x,labels=labels,autopct="%1.f%%")
# plt.legend(loc="upper left")
plt.legend(bbox_to_anchor=(1.0,1.0))
plt.show()
x
```
