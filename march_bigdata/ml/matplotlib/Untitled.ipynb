# Line Properties 

```python
# Line Properties
# linestyle : dotted(:),solid(-),dshed(--)
# line width /(lw)
# line size
# line color : color/c
```

```python
import numpy as np
a=np.array([10,2,15,3,6])#a=y axis values
plt.plot(a,marker="s",ms=20,mec="r",linestyle="--")
```

```python
import numpy as np
a=np.array([10,2,15,3,6])
plt.plot(a,marker="s",ms=20,mec="r",linestyle="--",lw=5,c="black")
```

```python
import numpy as np
x=[156,125,134,127,130,169]
y=[60,65,71,76,77,90]
plt.plot(x,y,marker="s",ms=5,mec="r",linestyle="-",lw=2.5,color="green")
plt.xlabel("Height in cm")
plt.ylabel("Weight in kg")
plt.title("Height-Weight")
```

```python
x=[156,125,134,127,130,169]
y=[60,65,71,76,77,90]
plt.plot(x,y,marker="s",ms=5,mec="r",linestyle="-",lw=2.5,color="green")
# json format passing font attributes for xlabel&ylabel
font1={"family":"serif","size":20,"color":"g"}
font2={"family":"sans-serif","size":10,"color":"c"}

plt.xlabel("Height in cm",fontdict=font2)
plt.ylabel("Weight in kg",fontdict=font2)
plt.title("Height-Weight",fontdict=font1)
```

```python
# x1,y1,x2,y2
# x1,y1
# x2,y2
x1=[10,15,20,25,30]
y1=[30,25,20,15,10]
x2=[5,7,9,11,12]
y2=[21,17,13,9,7]
plt.plot(x1,y1,marker="s",ms=5,mec="r",linestyle="-",lw=2.5,color="c")
plt.plot(x2,y2,marker="s",ms=5,mec="r",linestyle="-",lw=2.5,color="k")

# json format passing font attributes for xlabel&ylabel
font1={"family":"serif","size":20,"color":"g"}
font2={"family":"sans-serif","size":10,"color":"maroon"}

plt.xlabel("X",fontdict=font2)
plt.ylabel("Y",fontdict=font2)
plt.title("X-Y",fontdict=font1)
```
