**1)Find all of the numbers from 1-1000 that are divisible by 8**

```python
lst1=[ i for i in range(1,1001) if i%8==0]
lst1
```

**2)Find all of the numbers from 1-1000 that have a 6 in them**

```python
lst2=[i for i in range(1,1001) if "6" in str(i) ]
lst2
```

**4)Remove all the vowels in a string**

```python
lst4="".join([i for i in string if i not in "a,e,i,o,u" ])
lst4
```

**5)Find all the words in a string that are less than 5 letters**

```python
lst5=[i for i in string.split(" ") if 5>len(i)]
lst5
```
