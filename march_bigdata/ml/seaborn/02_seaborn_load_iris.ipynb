```python
import seaborn as sns
```

```python
df=sns.load_dataset("iris")
df
```

## Histogram 

```python
# histogram : univarient
sns.histplot(data=df,x="sepal_length")
```

```python
sns.histplot(data=df,x="petal_length")
```

```python
sns.histplot(data=df,x="petal_width")
```

```python
sns.histplot(data=df,x="sepal_width",color="bisque")
```

```python
# dist plot : combination of histplot and kde plot
```

## kde plot

```python
sns.kdeplot(df.sepal_length)
```

## Dist plot (Distribution Plot)
**dist plot : combination of histplot and kde plot**

```python
sns.displot(df.sepal_length,kde=True)
```

```python
sns.displot(df.sepal_length,kind="hist")
```

```python
sns.displot(df.sepal_length)
```

## Box plot 

```python
# used to find out outliers in dataset
```

```python
sns.boxplot(y="sepal_width",data=df)
# lower line is : lower outlier

#  25% - first quartile ( 1 dot)
#  50% quartile/median
#  75% - quartile

# upper line : upper outlier (3 dot)
```

```python
df.describe()
```

```python
# plot numerical values
sns.boxplot(df,palette="pastel")
```

```python
# showmeans display mean 
sns.boxplot(y="sepal_width",data=df,showmeans=True)
```

```python
# bivarient : point plot
```

## Point Plot 

```python
# identify growth of the data :virginica has higer values
# barplot/categorical
```

```python
sns.pointplot(x=df.species,y=df.sepal_length)
```

```python
# scatter plot
# bivariate analysis / numerical distribution
```

```python
sns.scatterplot(x=df.sepal_length,y=df.sepal_width)
```

## Line plot 

```python
# bivariate : Numerical plot
```

```python
sns.lineplot(x=df.sepal_length,y=df.sepal_width)
```

```python
sns.lineplot(x=df.petal_length,y=df.petal_width)
```

```python
# width increase - length also increase
```

## Regression 

```python
# regression : relation
# two features how they related
# independent variable mean &  dependent variable mean  relation
```

```python
sns.regplot(x=df.petal_length,y=df.petal_width)
# combination of dist plot and scatter plot
```

## Joint Plot

```python
# combination of regression plot and dist plot
```

```python
sns.jointplot(x=df.petal_length,y=df.petal_width)
```

```python
sns.jointplot(x=df.petal_length,y=df.petal_width,kind="reg")
# combination of dist plot and regression plot
```

```python
sns.jointplot(x=df.petal_length,y=df.petal_width,kind="hex")
# Kind : hex,scatter,reg,hist,kde
# reg is standard
```

## Pair Plot 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,kind="reg")
```

## Heat Map 

```python
# heatmap : to plot corelation
```

```python
df.corr(numeric_only=1)
```

```python
sns.heatmap(df.corr(numeric_only=1),cmap="crest")
```

```python
sns.heatmap(df.corr(numeric_only=1),cmap="mako",annot=True)
```
