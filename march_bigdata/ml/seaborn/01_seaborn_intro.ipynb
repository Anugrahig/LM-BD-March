```python
df=pd.read_csv("../../Bigdata Files/tips.csv")
df
```

```python
# count plot is univarient 
sns.countplot(x="smoker",data=df)
```

```python
sns.countplot(x="smoker",data=df,hue="sex")
```

```python
# time = count
sns.countplot(x="time",data=df,hue="day")
```

```python
sns.countplot(x="smoker",data=df,hue="sex",palette="PuBuGn")
```

```python
sns.countplot(x="smoker",data=df,hue="sex",palette="pastel")
# pastel
```
